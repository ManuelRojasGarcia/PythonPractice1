Construye un grafo dirigido con la librería `networkx` donde los nodos (etiquetados) representen a los diferentes emisores/receptores de los ataques y las aristas indiquen entre qué nodos se ha producido. <span style="fuente-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px; ">EI</span> **(1 punto)**

```python
#https://www.cienciadedatos.net/documentos/pygml01-introduccion-grafos-redes-python
#Para mostrar las flechas hay que cambiar de graph a digraph https://networkx.org/documentation/stable/reference/classes/digraph.html
# Más apuntes

import networkx as nx
import pandas as pd
import warnings
import matplotlib.pyplot as plt

attack = pd.read_csv("attack_sample.csv")

G = nx.DiGraph()
print(G)
G.is_directed()

G.add_nodes_from(attack["node.src"])
G.add_nodes_from(attack["node.dst"])

aristas= attack[["node.src", "node.dst"]].values
G.add_edges_from(aristas)

graph_pos = nx.spring_layout(G)

plt.figure(figsize=(5, 5))
nx.draw_networkx_nodes(G, graph_pos, node_size=300, node_color='red', alpha = 0.6)
nx.draw_networkx_labels(G, graph_pos, font_size=12, font_family='sans-serif')
nx.draw_networkx_edges(G, graph_pos, edge_color='black', alpha=0.5)


plt.title("Grafo de ataques dirigidos")
plt.axis('off')
plt.show()



```

```python
#Para total de ataques con in degree - https://networkx.org/documentation/stable/reference/classes/generated/networkx.DiGraph.in_degree.html
# Para contar nodos - https://stackoverflow.com/questions/28185802/counting-nodes-in-general-tree-in-python

import pandas as pd

# Leer el archivo CSV y cargar los datos en un DataFrame
attack = pd.read_csv("attack_sample.csv")
tiposvalores = attack ['node.src'].unique()


def countNodes(graph):
    return len(graph.nodes)

totalnodes = countNodes(G)
print("Los nodos son:",totalnodes, tiposvalores)

gradosentrada = [grado for nodo, grado in G.in_degree()]
print("Ataques recibido por tipo de nodo", gradosentrada)


valorgrado = gradosentrada[8]
print("El nodo más atacado ha sido", tiposvalores[8], "recibiendo un total de:", valorgrado, "ataques")

gradossalida = [grado for nodo, grado in G.out_degree()]
print("Ataques produdidos por tipo de nodo", gradossalida)

valorproducido = gradossalida[2]
print("El nodo que más ha producido ataques es", tiposvalores[2], "con un total de", valorproducido, "ataques")

print("Los nodos que no han recibido ningún ataque han sido", tiposvalores[1],tiposvalores[3],tiposvalores[6])

print("El número total de ataques es de:",sum(gradosentrada))
```

### Ejercicio Opcional

El [triángulo de Sierpinski](https://es.wikipedia.org/wiki/Triangle_de_Sierpinski) es un objeto fractal, que fue introducido por primera vez en 1915 por el matemático polaco Waclaw Sierpiński. El Tetraedro de Sierpiński es una generalización del fractal que se genera a partir de la subdivisión recursiva de un tetraedro. Empieza con un tetraedro inicial y después se divide repetidamente en tetraedros más pequeños, siguiendo un patrón específico. A cada paso de la subdivisión, se toman los puntos medio de las aristas para construir los nuevos tetraedros.

El resultado es un fractal en forma de tetraedros que se repiten a distintos niveles de detalle. Cada tetraedro más pequeño contiene una versión en miniatura del fractal completo. El Tetraedro de Sierpiński exhibe una estructura fractal autosimilar, con la misma forma que se repite en diferentes dimensiones y en una escala decreciente.

A medida que se incrementa el número de subdivisiones o niveles de detalle, el fractal resulta más complejo y muestra una intrincada estructura geométrica. Esta propiedad fractal, junto a su belleza visual y simetría, hace del Tetraedro de Sierpiński  una figura fascinante y cautivadora para los matemáticos y entusiastas de los fractales.

```python
# https://es.stackoverflow.com/questions/4649/obtener-y-gr%C3%A1ficar-triangulo-de-sierpinski
# El ejercicio se encuentra en stackoverflow, se comprenden los conceptos pero el contenido se obtiene de la web. Al ser el opcional lo incluyo aprendiendo del ejercicio.
# Incluyo algunas caracteristicas extras.

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

def get_triangle(P):
    m = P * 0.5
    n = P * 0.5 + np.array([0.5, 0])
    k = P * 0.5 + np.array([0.25, np.sqrt(3)/4])
    return np.array([m,n,k])

# triangulo unitario inicial
triangle = np.array([[0, 0],
              [1, 0],
              [0.5, np.sqrt(3)/2]])

# Crea un array que representa el triangulo Sierpinski hasta la etapa deseada
etapa = 5
for e in range(etapa):
    triangle = get_triangle(triangle)

# Se procede a graficar el triangulo
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')


for t in triangle.reshape(3**etapa,3,2):
    ax1.add_patch(mpatches.Polygon(t, fc="black"))


plt.axis('off')
plt.title("Triángulo de Sierpiński", fontsize=14, fontweight='bold')
plt.show()

foto =('/content/drive/MyDrive/Colab_Notebooks/prog_datasci_8/data/Waclaw_Sierpinski.jpg')
img = plt.imread(foto)
plt.imshow(img)
plt.axis('off')
plt.title("Waclaw Sierpiński")
plt.show()
```
